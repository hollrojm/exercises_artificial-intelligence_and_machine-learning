```python
import numpy as np
import scipy.linalg as linalg
A = np.array([[3,2], [1,4]])
b = np.array([6,5])

x = linalg.solve(A, b)
print("Solucion del sistema Ax = b:", x)
```

```python
from scipy.optimize import minimize
import numpy as np
def f(x):
    return x**2 + 5*np.sin(x)
result = minimize(f, x0=2.0)
print("Minimo encontrado en x = ", result.x)
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

camila = RandomForestClassifier()
camila.fit(X_train, y_train)
y_pred = camila.predict(X_test)


adriana = accuracy_score(y_test, y_pred)
print(f"Precision de modelo: {adriana:2f}")

nueva_flor = [[1.4, 1.4, 1.4, 1.4]]
prediccion = camila.predict(nueva_flor)
nombre_flor =iris.target_names[prediccion[0]]
print(f"La flor predicha es: {nombre_flor}")
```
